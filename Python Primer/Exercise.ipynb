### R-1.1
Write a short Python function, `is_multiple(n, m)`, that takes two integer values and returns True if `n` is a multiple of `m`, that is, `n = mi` for some integer `i`, and False otherwise.

```python
def is_multiple(n,m):
    return n%m==0

is_multiple(6,3), is_multiple(6,4)
```

#### R-1.2
Write a short Python function, `is_even(k)`, that takes an integer value and returns True if k is even, and False otherwise. However, your function
cannot use the multiplication, modulo, or division operators.

```python
def is_even(k):
    k = abs(k)
    if int(str(k)[-1]) in [0,2,4,6,8]:
        return True
    else:
        return False

is_even(345), is_even(348)
```

#### R-1.3
Write a short Python function, `minmax(data)`, that takes a sequence of
one or more numbers, and returns the smallest and largest numbers, in the
form of a tuple of length two. Do not use the built-in functions min or
max in implementing your solution

```python
def minmax(data):
    mn = data[0]
    mx = data[0]
    for i in data:
        if i < mn:
            mn = i
        if i > mx:
            mx = i
    return mn, mx
```

#### R-1.4
Write a short Python function that takes a positive integer n and returns
the sum of the squares of all the positive integers smaller than n.

```python
def sum_of_sqr(n):
    if n<=0:
        raise ValueError('n must be positive')
    sum_sqr = 0
    for i in range(n):
        sum_sqr += i*i
    return sum_sqr
```
